```
import glob #glob looks through directories and sub directories
import os
import shutil
import json
import csv
import zipfile #able to work inside a zip file without having to extract anything from the file
```

```
tweetzips = glob.glob('/content/drive/My Drive/Colab Notebooks/Network Analysis Project/*.zip') #axtris selects all of the zip files in the folder
```

```
# identify unique users in the mention network
uniqueusers = {}

for tweetzip in tweetzips:
  zf = zipfile.ZipFile(tweetzip)
  for i, obj in enumerate(zf.infolist()):
    tweetjson = json.load(zf.open(obj))
    userwhotweeeted = tweetjson['user']['screen_name']
    if userwhotweeeted in uniqueusers:
      uniqueusers[userwhotweeeted] += 1
    if userwhotweeeted not in uniqueusers:
      uniqueusers[userwhotweeeted] = 1 
```

```
userstoinclude =set()
# a set is a unique list
usercount = 0
for auser in uniqueusers:
  if uniqueusers[auser] > 2: #if any user has more than 1 tweet
    usercount += 1
    userstoinclude.add(auser)

print(len(userstoinclude))
```

```
edgelist = open('/content/drive/My Drive/Colab Notebooks/Network Analysis Project/new.edgelist.for.gephi.csv', 'w')
csvwriter = csv.writer(edgelist)
header =['Source', 'Target']
csvwriter.writerow(header)
```
